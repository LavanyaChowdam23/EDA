```python
# Bar plot of items per category
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Category', order=df['Category'].value_counts().index)
plt.title("Number of Menu Items per Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
#Average Calories per Category
avg_calories = df.groupby('Category')['Calories'].mean().sort_values(ascending=False)
avg_calories
```

```python
#Plot for Average Calories per Category
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_calories.values, y=avg_calories.index, palette="viridis")
plt.title("Average Calories per Category")
plt.xlabel("Calories")
plt.ylabel("Category")
plt.show()
```

```python
#cholestrol category
high_chol = df.groupby('Category')['Cholesterol (% Daily Value)'].mean()
high_chol
```

```python
#plotting based on cholestrol
plt.figure(figsize=(12, 6))
chol = df.groupby('Category')['Cholesterol (% Daily Value)'].mean().sort_values(ascending=False)
sns.barplot(x=chol.values, y=chol.index, palette="viridis")
plt.title("Average Cholesterol (% DV) per Category", fontsize=16)
plt.xlabel("Cholesterol (% DV)", fontsize=12)
plt.ylabel("Category", fontsize=12)
plt.show()
```

```python
#Top 5 Items by Nutrient
nutrients = ['Sugars','Dietary Fiber','Saturated Fat','Calories','Sodium','Cholesterol']
nutrient_data = df.groupby('Category')[nutrients].mean()
nutrient_data
```

```python
#plot the nutrient concentration per category
# Heatmap Visualization
plt.figure(figsize=(10, 6))
sns.heatmap(nutrient_data, annot=True, cmap="YlOrRd", fmt=".1f", linewidths=.5)
plt.title("Average Nutrient Content (% Daily Value) by Category", pad=20, fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

# Top Items for Each nutrient
for nutrient in nutrients:
    top_items = df.nlargest(5, nutrient)[['Category', 'Item', nutrient]]
    print(f"\nTop 5 {nutrient.split('(')[0]} Items:")
    print(top_items.to_markdown(index=False))
```

```python
#concentration of vitamins in each category
vitamins = ['Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)']
vitamin_data = df.groupby('Category')[vitamins].mean()
vitamin_data
```

```python
#plot the vitamin concentration per category
# Heatmap Visualization
plt.figure(figsize=(10, 6))
sns.heatmap(vitamin_data, annot=True, cmap="YlOrRd", fmt=".1f", linewidths=.5)
plt.title("Average Vitamin Content (% Daily Value) by Category", pad=20, fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
